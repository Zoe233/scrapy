## 第一节 期货及衍生品市场的形成与发展

### 考点一：期货及相关衍生品

#### 一、期货
1. 期货：不是货，是由**现货**衍生而来，是指以某种大宗商品或金融资产为标的可交易的标准化合约。

2. 期货合约：**期货交易所**统一制定的、定在将来某一特定的时间和地点交割的一定量标的物的标准化合约。

3. 期货合约包括：商品期货合约、金融期货合约、其他期货合约。
（根据标的物划分）

#### 二、远期
1. 远期：也称为远期合同或远期合约。

2. 远期合约：是指**交易双方**约定在未来的某一确定时间，以确定的价格买卖一定数量的某种标的资产的合约。

3. 常见的远期交易包括：
    - 商品远期交易
    - 远期利率协议FRA
    - 外汇远期交易
    - 无本金交割外汇远期交易NDF
    - 远期股票合约等

------
先有远期交易，才有期货。

远期是非标准化的。  
是否标准化的主要看合约是谁制定的。

A 和B 想要提前锁定未来的价格。  
A 和 B 签订远期合约。  
是由A，B自行商定签订的，所以不是标准化的，不能在市场中进行流通。  

远期合约具有非常大的信用风险。

#### 三、互换
1. 互换：指**两个或两个以上当事人**按照商定条件，在约定时间内交换一系列**现金流**的合约。远期合约可以看成仅交换一次现金流的互换。
2. 互换包括：利率互换、货币互换，商品互换、股权类互换和远期互换等。
3. 最常见也最重要的：是**利率互换**和**货币互换**。


#### 四、期权
1. 期权：一种**选择**的权利，即**买方**能够在未来的特定时间或者一段时间内按照事先约定的价格**买入或卖出**某种约定标的物的权利。
2. 按照标的资产划分，期权包括：利率期权、外汇期权、股权类期权和商品期权。
----
期权是买方拥有的选择的权利，卖方是不具有的。  
卖方负有相应的义务，在买方行驶权利时，卖方必须按照指定的价格买入或卖出。



期货和场内期权交易都是标准化的交易，标准化合约的场内交易。（期权也有场外交易市场交易的，非标准化合约）

远期和互换都是私人化的场外的交易。

### 考点二：现代期货市场的形成

#### 一、期货市场的萌芽
- 期货交易**最早萌芽于欧洲**。在古希腊和古罗马时期，欧洲出现中央交易场所和大宗易货交易，形成既定时间在固定场所开展的交易活动。
- 期货交易**萌芽于远期交易**。为期货交易奠定基础。  
远期交易是非标准化的。

#### 二、现代期货交易的形成
- 1848年 世界上第1家期货交易所**芝加哥期货交易所CBOT**成立
    - 这时仍然是私人化的非标准合约
    - 信用风险非常大
- 1865年 芝加哥期货交易所推出**标准化合约**，实行**保证金制度**
    - 保证金制度就是杠杆
    - 比如保证金是10%，1万元就能进行10万元的交易，帮助企业进行套期保值，规避风险
- 1882年 交易所允许以**对冲方式**履约责任
    - 对冲方式，在到期日之前，卖出平仓或买入平仓，了结了期货的投资行为
- 1883年 结算协会成立
- 1925年 芝加哥期货交易所结算公司（现代意义上结算机构）成立
    - 保管保证金，进行资金划拨，控制市场风险
   
   
标准化合约、保证金制度、对冲机制、统一结算的实施，标志着现代期货市场的确立。

#### 题目2：

- A 套利

- C 对冲

### 考点三：国内外期货市场发展趋势

#### 一、国际期货市场的发展历程
- 商品期货
    - 农产品期货
    - 金属期货：1876年，伦敦金融交易所LME成立
    - 能源化工期货
- 金融期货
    - 外汇期货：1972年CME设立国际货币市场分部
    - 利率期货：1975年CBOT推出世界第1个利率期货合约
    - 股指期货：1872年堪萨斯期货交易所开发出价格线综合指数期货合约
    - 股票期货：1995年，中国香港开始个股期货试点

#### 二、国际期货市场的发展趋势
- 国际期货市场的发展特点：
    - 交易中心日益集中
    - 交易所由**会员制**向**公司制**发展
    - 交易所**兼并重组**趋势明显
    - **金融期货**发展后来居上
    - 交易所竞争加剧，服务质量不断提高

#### 三、我国期货市场的发展历程
- 1. 初创阶段 1990-1993年
    - 1990年：郑州粮食批发市场引入期货交易机制
    - 1992年：第一家期货经纪公司（广东万通期货经纪公司）成立
    - 1993年：期货交易所达50多家，经纪机构达百家处于无序状态
- 2. 治理整顿阶段：1993-2000年
    - 1999年，期货交易所精简合并为3家（郑州商品交易所、大连商品交易所、上海期货交易所）
    - 2000年，期货业协会成立
    - 期货品种从35种下降至12种
- 3. 稳步发展阶段：2000-2013年
    - 2006年，中国期货保证金监控中心成立
    - 2006年，中国金融期货交易所成立
- 4. 创新发展阶段：2014年至今
    - 2014年国务院发布《关于进一步促进资本市场健康发展的若干意见》，简称“新国九条”：对资本市场改革发展进行**顶层设计和战略规划**。

这部分内容和书本的内容不一致。

#### 四、国内外远期、互换和期权市场的发展
- 远期市场
    - 布雷顿森林体系结束后
    - 1973年，外汇远期合约应运而生
- 互换市场
    - 1981年IBM与世界银行在伦敦签署世界上第1份利率互换协议
    - 2008年人民币利率互换交易开始正式全面推进
- 期权市场
    - 17世纪“荷兰郁金香”时期出现了最早的期权交易。
    - 1973年，CBOE推出第一张标准化期权合约

## 第二节 期货及衍生品主要特征

### 考点1：期货交易的基本特征
#### 一、期货交易的基本特征
1. 合约标准化；
    - 由期货交易所制定
2. 场内集中竞价交易；
3. 保证金交易；
4. 双向交易；
5. 对冲了解；
6. 当日无负债结算。

#### 题目4：
属于期货交易的基本特征的是（）
- A 无保证金交易
- B 当日无负债结算
- C 合约标准化
- D 场外交易

#### 答案：
BC

### 考点2：期货与远期的联系与区别
#### 一、联系
1. 二者均为未来时间买卖一定数量商品。是现货交易在时间上的延伸。
2. 期货交易是在远期交易的基础上发展起来的。

#### 二、区别
1. 交易对象不同：
    - 期货是标准化期货合约
    - 远期是交易双方私下协商达成的非标准化合同，所涉及的商品没有任何限制
2. 功能作用不同：
    - 期货（通过套期保值）规避风险和价格发现
    - 远期一定程度上调节供求关系，减少价格波动
3. 履约方式不同：
    - 期货是实物交割、对冲平仓
    - 远期是实物交割
4. 信用风险不同：
     - 期货较低
     - 远期较高
5. 保证金制度：
    - 期货保证金是合约价值的5%-15%
    - 远期是双方协商保证金数额

### 考点3：期货与期权的区别与联系
#### 一、联系
1. 期货合约和场内期权合约均是场内交易的标准化合约，均可以进行双向操作，均通过结算所统一结算；
2. 场内期权中又以期货期权与期货的联系最为密切，期货期权是期权和期货合约的有机结合。

#### 二、区别
1. 交易对象不同
    - 期货是标准化期货合约
    - 期权是买卖某种资产的权利
2. 权利义务
    - 期货合约是双向合约，交易双方都要承担期货合约到期交割的义务
    - 期权是单向合约，买方只享受权利，不承担义务
3. 保证金制度
    - 期货买卖双方都要缴纳期货合约价值5%-15%的保证金
    - 期权是买方不需要缴纳保证金，卖方缴纳一定的保证金
4. 盈亏特点
    - 期货交易的盈亏是对称的，交易双方都面临着无限的盈利和无止境的亏损，其盈亏曲线为线性。
    - 期权买方的收益不固定，亏损只限于购买期权的权利金；而卖方的收益最高不超过出售期权的权利金，亏损不固定。盈亏曲线是非线性的
5. 了解方式
    - 期货是对冲平仓，实物交割
    - 期权是对冲、行权和放弃权利

### 考点4：期货与互换的联系与区别
#### 一、联系
1. 远期合约是期货和互换的基础，期货和互换是对远期合约在不同方面创新后的衍生工具，互换利率与短期利率、期货利率联系密切；
2. 实践中常用期货给互换进行套期保值。

#### 二、区别
1. 标准化程度
    - 期货是标准化期货合约
    - 互换是非标准化合同，交易双方自行协商决定
2. 成交方式
    - 期货在有形的公开场内通过**电子交易**系统撮合成交
    - 无固定的交易场所和交易时间，主要通过**人工询价**的方式撮合成交
3. 合约双方关系
    - 交易者不关心交易对手的身份和信用
    - 交易双方都会对双方的信用和实力等方面做充分了解

## 第三节 期货及衍生品的功能和作用

### 考点1：期货及衍生品的功能
期货的功能：
- 规避风险
- 价格发现
- 资产配置

#### 一、 规避风险的功能
1. 实现过程：规避风险的功能是通过**套期保值**实现的。
2. 通过套期保值规避风险的原理
    - 对于同一种商品来说，在现货和期货市场同时存在情况下，在同一时空内会受到相同经济因素的影响和制约，两个市场的价格变动趋势相同。
    - 随着期货合约临近交割，现货价格与期货价格趋于一致。
3. 投机者的参与是套期保值实现的条件
    - 套期保值并不是消灭风险，只是将其转移，转移出去的风险需要有相应的承担着，期货投机者正是期货市场的**风险承担者**
    
#### 二、价格发现的功能
1. 原因：
    - 期货交易的参与者众多，大量的买方和卖方聚集在一起进行竞争，可以代表供求双方的力量，有助于公平价格的形成
    - 期货交易中的交易人士大多熟悉某种商品行情，有丰富的商品知识和广泛的信息渠道以及一套科学的分析、预测方法
    - 期货交易的透明度高，竞争公开化、公平化，有助于形成公正的价格
2. 特点
    - 预期性
    - 连续性
    - 公开性
    - 权威性

#### 三、资产配置的功能
1. 原因：
    - 借助期货能够为其他资产进行风险对冲
    - 期货市场的杠杆机制、保证金制度使得投资期货更加便捷和灵活，虽然风险较大，但同时也能获取高额的收益。
2. 原理：
    - 期货能够以套期保值的方式为现货资产或投资组合对冲风险，从而起到稳定收益、降低风险的作用
    - 商品期货是良好的保值工具
    - 将期货纳入投资组合能够实现更好的**风险--收益**组合

### 考点2：期货及衍生品市场的作用
#### 一、期货市场在宏观经济中的作用
1. 提供分散、转移价格风险的工具，有助于稳定国民经济
2. 为政府制定宏观经济政策提供参考依据
3. 促进本国经济的国际化
4. 有助于市场经济体系的建立与完善

#### 二、期货市场在微观经济中的作用
1. 锁定生产成本，实现预期利润
2. 利用期货价格信号，组织安排现货生产
3. 期货市场扩展现货销售和采购渠道
