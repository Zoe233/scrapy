## 第一节 期货交易所

### 考点1：期货交易所性质与职能
#### 一、性质
1. 期货交易所是为期货交易提供场所、设施、相关服务和交易规则的机构。它自身并不参与期货交易。
2. 是高度系统性和严密性、高度组织化和规范化的交易服务组织。

#### 二、职能
1. 提供交易的场所、设施和服务
2. 设计合约、安排合约上市
3. 制定并实施期货市场制度与交易规则
4. 组织并监督期货交易，监控市场风险
5. 发布市场信息

### 考点2：期货交易所组织形式
#### 组织形式
- 会员制
- 公司制

#### 一、会员制
1. 含义：是由**全体会员**共同出资组建，缴纳一定的会员资格费作为注册资本，以其全部财产承担有限责任的**非营利性法人**。
2. 会员制交易所组织架构
    - 会员大会：最高权力机构
    - 理事会：会员大会 的常设机构，对 会员大会 负责，执行 会员大会 决议。通过 会员大会 选举产生
    - 理事会专门委员会：由理事长提议，经理事会同意设立
    - 总经理：负责期货交易所日常经营管理工作的高级管理人员。设置相关业务部门，包括交易、交割、研究发展、市场开发、财务等部门 
3. 会员资格获取
    - 以交易所创办发起人的身份加入
    - 接受发起人的**资格转让**加入
    - 接受期货交易所其他会员的**资格转让**加入
    - 依据期货交易所的**规则加入**
4. 交易所会员权利
    - 参加会员大会，行使表决权、申诉权
    - 在期货交易所内进行期货交易，使用交易所提供的交易设施、获得期货交易的信息和服务
    - 按规定转让会员资格，联名提议召开临时会员大会等。
5. 交易所会员义务
    - 遵守国家法律、法规、规章和政策
    - 遵守期货交易所的章程、业务规则及有关决定
    - 按规定缴纳各种费用
    - 执行会员大会、理事会的决议
    - 接受期货交易所的业务监管等

#### 二、公司制
1. 含义：由若干**股东**共同出资组建，以**营利**为目的，股份可以按照有关规定转让，其盈利来自从交易所进行的期货交易中收取的各种费用。
2. 公司制交易所组织架构
    - 股东大会：全体股东共同组成，是公司制期货交易所的最高权力机构。对重大事项作出决议。
    - 董事会：公司制期货交易所的常设机构，执行股东大会决议
    - 监事会（或监事）：对股东大会负责，对公司财务以及公司董事、经理等高级管理人员履行职责的合法性进行**监督**
    - 总经理：负责期货交易所日常经营管理工作的高级管理人员。他对董事会负责，由**董事会聘任或解聘**。
3. 会员资格：在公司制期货交易所内进行期货交易，也必须获得会员资格


#### 三、公司制和会员制区别
- 经营目标
    - 会员制：不以营利为目标
    - 公司制：以营利为目标，追求利润最大化
- 适用法律：
    - 会员制：适用民法的有关规定
    - 公司制：优先适用公司法规定，在公司法未做出规定情况下适用民法
- 决策机构：
    - 会员制：会员大会，常设机构是 理事会
    - 会员制：股东大会，常设机构是 董事会

### 考点3：我国境内期货交易所
- 境内交易所
    - 会员制
        - 郑州商品交易所
        - 上海期货交易所
        - 大连商品交易所
    - 公司制
        - 中国金融期货交易所（2006年成立）
#### 一、境内交易所会员管理
1. 取得期货交易所会员资格，应经期货交易所批准。
2. 期货交易所对会员实行**总数控制**，只有会员才能取得场内交易席位，非会员须通过期货公司代理交易。
3. 会员制交易所会员权利：
    - 参加会员大会，行使选举权、被选举权和表决权；
    - 在期货交易所从事规定的交易、结算和交割业务；
    - 使用期货交易所提供的交易设施，获得期货交易的信息和服务，按规定转让会员资格；
    - 联名提议召开临时会员大会；
    - 按照期货交易所章程行使申诉权。
4. 会员制交易所会员义务：
    - 遵守法律法规
    - 遵守交易所章程、交易规则
    - 按规定缴纳各种费用
    - 执行会员大会、理事会的决议
    - 接受期货交易所监督管理
5. 公司制交易所会员权利：
    - 在期货交易所从事规定的交易、结算和交割等业务
    - 使用交易所提供的交易设施、信息和服务
    - 按交易规则行使申诉权
6. 公司制交易所会员义务：
    - 遵守法律法规
    - 遵守期货交易所章程、交易规则和实施细则
    - 按规定缴纳各种费用
    - 接受期货交易所监督管理

## 第二节 期货结算机构

### 考点1：期货结算机构性质与智能
#### 一、性质
负责交易所**期货交易的统一结算、保证金管理和结算风险控制**的机构。

#### 二、职能
1. 担保交易履约
2. 结算交易盈亏
3. 控制市场风险

### 考点2：期货结算结构的组织形式
#### 一、是某一交易所的内部机构。
仅为该交易所提供结算服务。我国采取此形式。
#### 二、是独立的结算公司。
可为一家或多家期货交易所提供结算服务。

### 考点3：期货结算制度（三个层次）
- 第一个层次：
由**结算机构对结算会员**进行结算，结算会员是交易所会员中资金雄厚、信誉良好的期货公司或金融机构。
- 第二个层次：
**结算会员与非结算会员或者结算会员与结算会员所代理客户**之间的结算。
- 第三个层次：
**非结算会员对非结算会员所代理客户**的结算。

### 考点4：我国境内期货结算机构
我国境内四家期货交易所的结算机构：**均是交易所的内部机构**，期货交易所既提供交易服务，也提供结算服务。  

我国境内期货结算制度：分为**全员结算制度**和**会员分级结算制度**两种类型。

#### 1. 全员结算制度
- 指期货交易所会员均具有与期货交易所进行结算的资格，期货交易所的会员均既是交易会员，也是结算会员，没有结算会员与非结算会员之分。
- 期货交易所对会员进行结算，会员对其受托的客户结算。
- 郑州商品交易所、大连商品交易所和上海期货交易所实行全员结算制度。

#### 2. 会员分级结算制度
- 期货交易所将交易所会员区分为结算会员与非结算会员。
    - 结算会员具有与交易所进行结算的资格；
    - 非结算会员不具有与期货交易所进行结算的资格
- 期货交易所对结算会员结算，结算会员对非结算会员结算，非结算会员对其委托的客户结算。
- 中国金融期货交易所：会员分级结算制度。

## 第三节 期货中介与服务机构

### 考点1：期货公司
#### 一、期货公司的性质、特征、职能
1. 期货公司的性质
    - 期货公司
        - 属于非银行金融机构。
            - 银行以外的各种经营金融业务的信用机构（保险公司、期货公司、信用合作社、消费信用机构、信托投资公司）
        - 是代理客户进行期货交易并收取交易佣金的中介组织。
        - 是场外期货交易者与期货交易所之间的桥梁和纽带。
        

2. 期货公司特征
    - 期货公司是依托从商品、资本、货币市场等衍生出来的市场提供风险管理服务的**中介机构**。
    - 客户的**保证金风险**往往成为期货公司的重要风险源。
    - 期货公司高度重视客户利益，面临**双重代理关系**，即公司股东与经理层的委托代理问题以及客户与公司的委托代理问题。

3. 期货公司职能
    - 按客户指令**代理**买卖期货合约、办理结算和交割手续
    - 对客户**账户**进行管理，控制客户交易风险
    - 为客户提供期货**市场信息**，进行期货交易咨询，充当客户交易顾问
    - 为客户**管理资产**，实现财务管理
