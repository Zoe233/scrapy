## 第一节 期货投机交易

### 考点一：期货投机
#### 一、含义
**期货投机**：指交易者通过预测期货合约**未来价格**的变化，以在期货市场上**获取价差**收益为目的的期货交易行为。  

#### 二、期货投机与套期保值区别
|项目|期货投机|套期保值交易|
|:-|:-|:-|
|交易目的|赚取**价差**|**规避**现货价格波动**风险**|
|交易方式|**单方向**买空卖空，获得价差收益|在**现货**市场与**期货**市场**同时**操作|
|交易风险|通常是为**博取价差**收益而主动承担相应价格风险，是价格风险偏好者|通过期货交易**规避**现货价格风险，是价格风险厌恶者|

#### 三、期货投机者类型
1. 按照交易**主体**划分：
机构投机者和个人投机者
2. 按**持有头寸**方向划分
    - **多头**投机者：预测价格上涨买进期货合约
    - **空头**投机者：预测价格下跌卖出期货合约



### 考点2：期货投机操作方法
#### 一、开仓阶段
1. 入市时机的选择
2. 金字塔式建仓
3. 合约交割月份的选择


1. 入市时机的选择
    - 通过基本分析法，**判断市场**处于牛市还是熊市，然后进一步利用技术分析升跌势有多大，持续时间有多长。
    - 权衡风险和获利前景。
    - 决定入市的具体时间。  
    
2. 金字塔式建仓
    - 原理：如果建仓后市场行情走势与预期相同并已使投机者获利，可增加持仓。
    - 原则：
        - 只有在现有持仓已盈利的情况下，才能增仓；
        - 持仓的增加应渐次递减。
    - 特点：将不断**买入（卖出）**的期货合约的平均价格保持在**较低（高）**水平。
    
    - 金字塔式建仓——案例
        - 交易单位：10吨/手 
        
|价格（元/吨）|持仓数（手）|平均价（元/吨）|
|:-|:-:|:-|
|2050|0|2026.3|
|2040|0 0 | 2024.6|
|2030|0 0 0 | 2022|
|2025|0 0 0 0| 2019.4|
|2015|0 0 0 0 0| 2015|

3. 合约交割月份选择
    - 选择合约的交割月份时，注意合约的**流动性**；远月合约价格与近月合约价格之间的关系。  
    - 期货投机者在选择合约月份时，应选择交易活跃的合约月份，**避开不活跃的合约月**。  
    - **正向市场**中，远月合约的**价格高于**近月合约的价格。多头投机者应买入近月合约；空头投机者应卖出远月合约。
    - **反向市场**中，远月合约的**价格低于**近月合约的价格。做**多头**的投机者宜买入皎月月份较远的远月合约；而做**空头**的投机者宜卖出交割月份较近的近月合约。  
    
    
#### 二、平仓阶段
1. 投资者建仓后应密切关注行情的变动，**适时平仓**。   
行情变动有利时，通过平仓获取投机利润；行情变动不利时，通过平仓可以限制损失。 

2. 投机者应灵活运用**止损指令**实现限制损失、累积盈利。   
**止损指令**是实现“限制损失、累积盈利”的有力工具。

## 第二节 期货套利交易

### 考点1：期货套利及作用
#### 一、期货套利
期货套利：分为**价差套利**和**期现套利**。  
1. 价差套利：利用期货市场上**不同合约之间**的价差进行的套利行为。价差套利有助于不同期货合约价格之间的合理价差关系的形成，提高市场流动性。 

2. 期现套利：利用**期货市场与现货市场**之间的不合理价差，通过在两个市场**反**向交易，待价差趋于合理而获利的交易。  


#### 二、期货套利作用
1. 期货价差套利行为有助于不同期货合约价格之间的**合理价差关系**的形成。
2. 期货价差套利行为有助于**提高市场流动性**。  



### 考点2：价差与期货价差套利
#### 一、期货价差
期货价差：期货市场上两个不同月份或不同品种期货合约之间的价格差。 
1. 建仓时价差：用价格较**高**的一“边”减去价格较**低**的一“边”。
2. 平仓时价差：为保持一致性，也要用**建仓时**价格**较高**的合约平仓价格减去建仓时价格**较低**的合约平仓价格。 

#### 二、价差扩大与缩小
1. 如果**平仓时**价差**大于建仓时**价差，则价差是**扩大**的；
2. 如果**平仓时**价差**小于建仓时**价差，则价差是**缩小**的。

#### 三、价差变动与套利盈亏计算
可**分别**计算每个期货合约的盈亏，然后进行**加总**，得到整个套利交易的盈亏。 

#### 四、价差扩大与买入套利
如果套利者**预期**两个或两个以上期货合约的**价差将扩大**，则套利者将**买入**其中**价格较高**的合约，**同事卖出价格较低**的合约，我们称这种套利为**买入套利**（Buy Spread）。如果价差变动方向与套利者的预期相同，则套利者同时将两份合约平仓而获利。  

#### 五、价差缩小与卖出套利
如果套利者**预期**两个或两个以上相关期货合约的**价差将缩小**，套利者可通过**卖出**其中价格**较高**的合约，同时**买入**价格的合约进行套利，我们称这种套利为**卖出套利**（Sell Spread）。

#### 六、期货价差套利指令
1. 价差套利市价指令
在使用这种指令时，套利者不需要注明价差的大小，只需注明买入和卖出期货合约的种类和月份即可；具体成交的价差如何，则取决于指令执行时点上市场行情的变化情况。

    - 优点：成交速度快
    - 缺点：在市场行情发生较大变化时，成交的价差可能与交易者最初的意图有较大差距。

2. 套利限价指令
若套利者希望以一个**理想的价差成交**，可使用套利限价指令。指令将以指定的或更优的价差来成交。

    - 优点：保证交易者以理想的价差进行套利
    - 缺点：使用该指令不能保证立刻成交

## 第三节 期货套利的基本策略

### 考点1：跨期套利
跨期套利，在**同一市场**（交易所）同时买入、卖出**同一期货品种**的**不同交割月份**的期货合约，以期在有利时机同时将这些期货合约对冲平仓获利。   
在实际操作中，根据套利者对不同合约月份中近月合约与远月合约买卖方向的不同，跨期套利可分为**牛市套利、熊市套利和蝶式套利**。 

#### 一、牛市套利
当市场**供给不足、需求旺盛**或远期供给相对旺盛的情形，导致较近月份合约价格上涨幅度大于较远月份合约价格的上涨幅度，或者较近月份合约价格下降幅度小于较远月份合约价格的下跌程度，**无论是正向市场还是反向市场**，此时，**买入较近**月份的合约**同时卖出较远月份**的合约进行套利，**盈利**的可能性比较大。 

1. 正向市场，牛市套利者**买入较近月份**的合约，同时**卖出较远月份**合约，为**卖出套利**的情形，只有在两个合约**价差缩小**时才能够盈利。

2. 反向市场，牛市套利者**买入较近月份**的合约，同时**卖出较远月份**合约，为**买入套利**的情形，只有在**两个合约价差扩大**时才能盈利。  

3. 正向市场，牛市价差缩小盈利；反向市场，牛市价差扩大盈利。 

#### 二、熊市套利
市场出现供给过剩，需求相对不足时，一般**较近月份的合约价格下降幅度**往往**要大于较远期合约价格的下降幅度**，或**较近月份的合约价格上升幅度较远合约价格的上升幅度**。无论是正向市场还是在反向市场，卖出较近月份的合约同时买入远期月份的合约进行套利，盈利的可能性比较大，称为**熊市套利**。  

1. 正向市场，熊市套利者**卖出较近月份**的合约，同时**买入较远月份**的合约进行套利，属于**买入套利**的情形，则只有两个合约**价差扩大**时才能够盈利。

2. 反向市场，熊市套利交易者**卖出较近月份**的合约，同时**买入较远月份**的合约进行套利，属于**卖出套利**的情形，则只有两个合约**价差缩小**时才能够盈利。

3. 正向市场，熊市价差扩大盈利；反向市场，熊市价差缩小盈利。  

#### 三、蝶式套利
1. 蝶式套利：由共享居中交割月份**一个牛市套利**和**一个熊市套利**的跨期套利组合。由于较近月份和较远月份的期货合约分别处于居中月份的**两侧**，形同蝴蝶的两个翅膀，因此称之为蝶式套利。 


2. 蝶式套利操作方法：买入（或卖出）**较近**月份合约，**同时卖出**（或买入）**居中**月份合约，**并买入**（或卖出）**较远**月份合约，其中，**居中月份**合约的**数量等于**较近月份和较远月份数量**之和**。 是在近期与居中月份之间的牛市（或熊市）套利和在居中月份与远期月份之间的熊市（或牛市）套利的一种组合。 




**卖出套利**的关键在于**价差缩小**才能盈利。  

### 考点2：跨品种套利
**跨商品套利**：利用两种或三种不同的但**相互关联的商品**之间的期货合约价格差异进行套利，即同时买入或卖出某一交割月份的相互关联的商品期货合约，以期在有利时机同时将这些合约对冲平仓获利。

两种情况：
- 一是相关商品之间的套利；
- 二是原材料与成品之间的套利。

#### 一、相关商品之间的套利
相关联商品之间的比值关系经常偏离合理的区间，**表现**为一种商品被**高估**，另一种**被低估**，或相反，从而为跨品种套利带来了可能。在此情况下，交易者可以通过期货市场**卖出被高估**的商品合约，**买入被低估**的商品合约进行套利，等有利时机出现后分别**平仓**，从中获利。  

#### 二、原材料与成本之间的套利
1. 大豆提油套利   
做法：购买大豆期货期货合约的同时，卖出豆油和豆粕的期货合约，当在现货市场上购入大豆或将成品最终销售时再将期货合约对冲平仓。  
这样，大豆加工商就可以锁定产成品和原料间的价差，防止市场价格波动带来的损失。 


2. 反向大豆提油套利   
做法：卖出大豆期货合约，买进豆油和豆粕期货合约，同时缩减生产，减少豆粕和豆油的供给量，三者之间的价格将会趋于正常，大豆加工商在期货市场中的盈利将有助于弥补现货市场中的亏损。  




### 考点3：跨市套利
#### 一、含义
跨市套利：也称市场间套利，是指在某个交易所买入（或卖出）某一交割月份的某种商品合约的同时，在另一个交易所卖出（或买入）同一交割月份的同种商品合约，以期在有利时机分别在两个交易所同时对冲所持有的合约而获利。 

- 品种相同
- 期限相同
- 市场（交易所）不同

### 考点4：期现套利
#### 一、期现套利
期现套利：是通过利用**期货**市场和**现货**市场的不合理价差进行反向交易而获利。 

#### 二、期现套利操作情形
1. 若价差远远**高于持仓费**，套利者就可通过**买入现货**，同时卖出相关期货合约，待合约到期时，用所买入的现货进行交割。
2. 若价差远远**低于持仓费**，套利者就可通过**卖出现货**，同时买入相关期货合约，待合约到期时，用交割获得的现货来补充之前所卖出的现货。
